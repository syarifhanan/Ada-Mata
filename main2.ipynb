```python
# Train YOLOv11s Model

# Import YOLO from ultralytics
from ultralytics import YOLO

# Load a model
model = YOLO('yolo11n.pt')  # load a pretrained model (recommended for training)

# Train the model
history = model.train(data='dataset/dataset.yaml', 
            epochs=60, 
            imgsz=640,
            flipud=0.5, # data augmentation flip up-down
            fliplr=0.5, # data augmentation flip left-right
            degrees=90) # data augmentation rotation degrees
```

```python
# Test the Model

# Load the trained model
model = YOLO('runs/detect/train2/weights/best.pt')

# Run prediction on the validation images
results = model.predict(source='dataset/images', save=True)
```
